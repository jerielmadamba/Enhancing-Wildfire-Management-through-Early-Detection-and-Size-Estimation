### 1.1 View the dataset


#### Data Dictionary

**fire_year:** The wildfire year runs from 00:00 hours on April 1 until 23:59 hours on March 31. The fire year is named for the year
in which April 1 falls. For example, the 2013 fire year begins April 1, 2013 and ends March 31, 2014.

**fire_number:** The first letter of the wildfire number indicates the forest area in which the ignition area is located. Three digits (with leading zeros) number wildfires consecutively within each forest area starting April 1 and
terminating March 31 of each year.

**fire_name:** Wildfires of provincial significance are assigned a geographic name for ease of recognition. For example: Mosquito Lake Fire.

**current_size:** This is the final area burned of the wildfire. This size will not necessarily match the sizes from the fire status fields. If
an updated size is determined after a wildfire has already been extinguished, the revised size is recorded in the Area Burned field of FIRES and the extinguished size recorded on the fire status line will not change.

**size_class:** This is a grouping of wildfires into classes based on final area burned. Size class can change as the wildfire grows until the final area burned is determined after extinguishment.<br>
A class = 0 to 0.1 ha<br>
B class > 0.1 ha to 4.0 ha<br>
C class > 4.0 ha to 40.0 ha<br>
D class > 40.0 ha to 200 ha<br>
E class > 200 ha

**fire_location_latitude and  fire_location_longitude:** Location of the ignition area of the wildfire recorded as a GPS (recorded as decimal degrees (DD).

**fire_origin:** This indicates who owns or administers the land the wildfire is burning on. If the wildfire covers more than one origin, this will be the land where the ignition area of the wildfire is.

**general_cause_description:** Classification of the wildfire cause according to the general group, individual industry or ignition source (for lightning)
that started the wildfire.

**industry_identifier_desc:** Two of the above mentioned general cause categories require further detail through the industry identifier Government or Other Industry.

**responsible_group_desc:** For the general cause of Recreation, identify the recreational group responsible for causing the wildfire. (campers (includes day use), hunters, other)

**activity_class:** Identifies the specific activity that was going on when the wildfire was started.

**true_cause:** Identifies the specific reason why the wildfire started.

**fire_start_date:** The time and date the wildfire started.

**det_agent:** A more specific description of the detection agent that discovered the wildfire.

**discovered_date:** The time the detection agent first discovered the wildfire.

**det_agent_type:** The general type of detection agent responsible for discovering the wildfire.

**discovered_size:** The size of wildfire when first discovered by the detection agent.

**reported_date:** The time and date the wildfire was first reported to Wildfire Management. This is generally the time the discovery agent
calls the Area dispatch.

**dispatched_resources:** The first Wildfire Management resource that started for the wildfire

**dispatched_date:** The date and time the first Wildfire Management resource was dispatched to the wildfire.

**start_for_fire_date:** The time and date that the first Wildfire Management resource started for the wildfire.

**assesment_resource:** This is the name (first and last) of the Wildfire Management person who first assessed the wildfire. This assessment could
be a quick assessment or the full assessment on the FP41.

**assessment_datetime:** The date and time the wildfire was originally assessed by Wildfire Management employees.

**assessment_hectares:** The size of the wildfire at the time of assessment is recorded to the nearest hectare or hundredth (0.01) of a hectare.

**fire_spread_rate:** The rate of spread of the wildfire at the time of initial assessment, capture in metres per minute.

**fire_type:** This is the predominant fire behaviour that was observed when the fire was originally assessed (ground, surface, crown).

**fire_position_on_slope:** The position of the wildfire relative to the slope it is travelling on at the time of initial assessment. (Bottom Flat, Lower 1/3, Middle 1/3, Upper 1/3).

**weather_conditions_over_fire:** Description of the weather conditions over the wildfire at the time of initial assessment. (CB dry, CB wet, Clear Cloudy, Rain, showers)

**temperature:** The temperature at the wildfire site at the time of initial assessment, captured in degrees Celsius.

**relative_humidity:** The relative humidity at the wildfire site at the time of initial assessment.

**wind_direction:** The wind direction at the wildfire site at the time of initial assessment.

**wind_speed:** The wind speed at the wildfire site at the time of initial assessment, captured in kilometers per hour.

**fuel_type:** The predominate fuel type (vegetation cover) in which the wildfire is burning, at the time of initial assessment.

**initial_action_by:** The resource that took the first physical suppressive action directly on the wildfire.

**ia_arrival_at_fire_date:** This is the date and time that the initial action resource (first physical suppression) arrived at the wildfire.

**ia_access:** This is only completed and captured in FIRES for wildland firefighting crews (HAC, UNIT, FTAC). All resources can record
this information on the FP41.

**fire_fighting_start_date:** The time and date that the initial action resource began physical suppression on the wildfire.

**fire_fighting_start_size:** The size of the wildfire at the time the initial action resource began physical suppression, recorded to the nearest hectare or hundredth (0.01) of a hectare.

**bucketing_on_fire:** If a helicopter with a bucket was used during the initial action of the wildfire, choose ‘yes’.

**distance_from_water_source:** If a helicopter with a bucket was used during initial action of the wildfire, record the distance in kilometers (to the nearest tenth) that the rotor wing flew from the water source to the wildfire.

**first_bucket_drop_date:** If a helicopter with a bucket was used during initial action of the wildfire, record the date and time of the first bucket drop on the wildfire.

**bh_fs_date:** The date and time the Incident Commander (IC) determined the wildfire to be at the fire status of being held (BH).

**bh_hectares:** The size of the wildfire at the time the Incident Commander changed the status to Being Held (BH), recorded to the nearest hectare or hundredth (0.01) of a hectare.

**uc_fs_date:** The date and time the Incident Commander determined the wildfire to be at the fire status of under control (UC).

**uc_hectares:** The size of the wildfire at the time the Incident Commander changed the status to Under Control (UC), recorded to the nearest hectare or hundredth (0.01) of a hectare.

**to_fs_date:** This is the date and time the Incident Commander determined the wildfire to be at the fire status of turned over (TO).

**to_hectares:** The size of the wildfire at the time the Incident Commander turned over the extinguishment of the wildfire to another agency or individual, recorded to the nearest hectare or hundredth (0.01) of a hectare.

**ex_fs_date:** This is the date and time the Incident Commander determined the wildfire to be completely extinguished (EX).

**ex_hectares:** The size of the wildfire at the time the Incident Commander changed the status to Extinguished (EX), recorded to the nearest hectare or hundredth (0.01) of a hectare.

#### 2.3 Impute each attributes missing values

```python
# Count of all missing data records
print("The dataset have",wildfire.isnull().sum().sum(), "missing records.")
```

```python
# Count of missing data records for each attribute
wildfire.isnull().sum()
```

I will use the mode for categorical attributes and the mean for numerical attributes during imputation, to ensure that the "fire_start_date" remains the earliest among other datetime attributes, I will employ frontfilling for this particular attribute. However, for the remaining attributes, I will use backfilling as the appropriate method to handle missing values. This approach will maintain the dataset's chronological order while filling in the null values effectively.

```python
# Fire Origin - mode
wildfire['fire_origin'].fillna(wildfire['fire_origin'].mode().iloc[0], inplace=True)

# Fire Start Date - backfill
wildfire['fire_start_date'] = wildfire['fire_start_date'].ffill()

# Det Agent Type - The general type of detection agent responsible for discovering the wildfire. - mode
wildfire['det_agent_type'].fillna(wildfire['det_agent_type'].mode().iloc[0], inplace=True)

# Discovered Date - backfill
wildfire['discovered_date'] = wildfire['discovered_date'].bfill()

# Reported Date - backfill
wildfire['reported_date'] = wildfire['reported_date'].bfill()

# Dispatched Resource - mode
wildfire['dispatched_resource'].fillna(wildfire['dispatched_resource'].mode().iloc[0], inplace=True)

# Dispatched Date - backfill
wildfire['dispatch_date'] = wildfire['dispatch_date'].bfill()

# Assessment Resource - mode
wildfire['assessment_resource'].fillna(wildfire['assessment_resource'].mode().iloc[0], inplace=True)

# Wildfire Management resource started for the wildfire
wildfire['start_for_fire_date'] = wildfire['start_for_fire_date'].bfill()

# Fire Spread Rate - mean
wildfire['fire_spread_rate'].fillna(wildfire['fire_spread_rate'].mean(), inplace=True)

# Fire Type - mode
wildfire['fire_type'].fillna(wildfire['fire_type'].mode().iloc[0], inplace=True)

# Fire Position on Slope - mode
wildfire['fire_position_on_slope'].fillna(wildfire['fire_position_on_slope'].mode().iloc[0], inplace=True)

# Weather conditions over fire - mode
wildfire['weather_conditions_over_fire'].fillna(wildfire['weather_conditions_over_fire'].mode().iloc[0], inplace=True)

# Temperature - mode
wildfire['temperature'].fillna(wildfire['temperature'].mode().iloc[0], inplace=True)

# Relative Humidity - mean
wildfire['relative_humidity'].fillna(wildfire['relative_humidity'].mean(), inplace=True)

# Wind Direction - mode
wildfire['wind_direction'].fillna(wildfire['wind_direction'].mode().iloc[0], inplace=True)

# Wind Speed - mean
wildfire['wind_speed'].fillna(wildfire['wind_speed'].mean(), inplace=True)

# Fuel Type - mode
wildfire['fuel_type'].fillna(wildfire['fuel_type'].mode().iloc[0], inplace=True)

# Initial Action By - mode
wildfire['initial_action_by'].fillna(wildfire['initial_action_by'].mode().iloc[0], inplace=True)

# Fire Fighting Start Date - backfill
wildfire['fire_fighting_start_date'] = wildfire['fire_fighting_start_date'].bfill()

# Fire Start Date when the fire was under control - backfill
wildfire['uc_fs_date'] = wildfire['uc_fs_date'].bfill()

# Hectares burned when declared under control - mena
wildfire['uc_hectares'].fillna(wildfire['uc_hectares'].mean(), inplace=True)
```

### 2.5 Convert datetime attributes to numerical

By calculating the time intervals between each attribute and the "fire_start_date," we convert them to numerical form. This transformation allows us to utilize time-related information more effectively.

```python
# discovered_date
# Due to imputation, some value may result to negative values,
# therefore we transform them to 0 and treat the date as the same day as the fire start date.
discover = (wildfire['discovered_date'] - wildfire['fire_start_date']).dt.days
# if less than 0, set as 0
wildfire['imp_discovered_date'] = discover.where(discover >= 0, 0)

# reported_date
report = (wildfire['reported_date'] - wildfire['discovered_date']).dt.days
# treat negative values the same as the preceding date, and assume that both scenario happend on the same day
wildfire['imp_reported_date'] = report.where(report >= 0, wildfire['imp_discovered_date'])

# dispatch_date
dispatch = (wildfire['dispatch_date'] - wildfire['reported_date']).dt.days
# treat negative values the same as the preceding date, and assume that both scenario happend on the same day
wildfire['imp_dispatch_date'] = dispatch.where(dispatch >= 0, wildfire['imp_reported_date'])

# start_for_fire_date
# this date is similar to dispatch_date, therefore dropping it.
wildfire = wildfire.drop(columns=['start_for_fire_date'])

# assessment_datetime
# This date only determines the fire year in which the wildfire occurs.
# therefore we will be dropping it since we will be using fire_start_date for the fire year.
wildfire = wildfire.drop(columns=['assessment_datetime'])

# fire_fighting_start_date
fire_fighting = (wildfire['fire_fighting_start_date'] - wildfire['dispatch_date']).dt.days
# treat negative values the same as the preceding date, and assume that both scenario happend on the same day
wildfire['imp_fire_fighting_start_date'] = fire_fighting.where(fire_fighting >= 0, wildfire['imp_reported_date'])

# bh_fs_date
being_held = (wildfire['bh_fs_date'] - wildfire['fire_fighting_start_date']).dt.days
# treat negative values the same as the preceding date, and assume that both scenario happend on the same day
wildfire['imp_bh_fs_date'] = being_held.where(being_held >= 0, wildfire['imp_fire_fighting_start_date'])

# uc_fs_date
under_control = (wildfire['uc_fs_date'] - wildfire['bh_fs_date']).dt.days
# treat negative values the same as the preceding date, and assume that both scenario happend on the same day
wildfire['imp_uc_fs_date'] = under_control.where(under_control >= 0, wildfire['imp_bh_fs_date'])
```

Allocate the imputed attributes to their respective correct names. Then drop the imputed categories. Then remove the imputed attributes and convert them into a numerical data type.

```python
update_col = ['discovered_date', 'reported_date', 'dispatch_date',
                'fire_fighting_start_date', 'bh_fs_date', 'uc_fs_date']
for column in update_col:
    imputed_column = f'imp_{column}'
    wildfire[column] = wildfire[imputed_column]
    wildfire = wildfire.drop(columns=imputed_column)
    wildfire[column].astype(int)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))
sns.histplot(wildfire['fire_spread_rate'], bins=10, ax=axes[0])
plt.boxplot(wildfire['fire_spread_rate'],patch_artist = True)

axes[0].set_title('Distribution of Fire Spread Rate')
axes[1].set_title('Range of the Fire Spread Rate')
plt.tight_layout()
plt.show()
```

The distribution of the fire spread rate is  right skewed. This means that the majority of the fire spread rate values are concentrated towards the lower end, while a few extreme values extend the right tail of the distribution.

```python
# calculating Q1 for fire_spread_rate attribute
# 0.25 means 25% on left and 75% on right
Q1_fire_spread_rate = wildfire['fire_spread_rate'].quantile(0.25)
print("Q1_fire_spread_rate:", float(Q1_fire_spread_rate),"%")
# calculating Q3 for fire_spread_rate attribute
Q3_fire_spread_rate = wildfire['fire_spread_rate'].quantile(0.75)
print("Q3_fire_spread_rate:", float(Q3_fire_spread_rate),"%")
# calculate interquartile (IQR) for fire_spread_rate attribute
IQR_fire_spread_rate = Q3_fire_spread_rate - Q1_fire_spread_rate
print("IQR_fire_spread_rates", float(Q3_fire_spread_rate),"%")
#calculating lower bound for fire_spread_rate atrribute
lowerBound_fire_spread_rate= Q1_fire_spread_rate - (1.5 * IQR_fire_spread_rate)
lowerBound_fire_spread_rate
#calculating upper bound for fire_spread_rate atrribute
upperBound_fire_spread_rate = Q3_fire_spread_rate + (1.5 * IQR_fire_spread_rate)
upperBound_fire_spread_rate
```

```python
#removing the outlier
index = wildfire[(wildfire['fire_spread_rate']>upperBound_fire_spread_rate)].index
wildfire = wildfire.drop(index)
index = wildfire[(wildfire['fire_spread_rate']<lowerBound_fire_spread_rate)].index
wildfire = wildfire.drop(index)
```

## **6. Feature Engineering**

```python
# Dropping latitude and longitude as they are unnecessary for our modeling purposes, and fire start date as it has been substituted with fire start year and fire start month.
wildfire = wildfire.drop(columns=['fire_location_latitude','fire_location_longitude','fire_start_date'])
```

Get the predictor attributes.

```python
predictors = wildfire.drop(columns='size_class')
```

Count the unique records for each attributes

```python
# Get all categorical columns
category_data = predictors.select_dtypes(include='category')

for col in category_data.columns:
    unique_count = category_data[col].nunique()
    print(f"{col} = {unique_count}")
```

Use Label Encoding for attributes that have equal or more than 8 unique records, and One Hot Encoding for attributes that have less than 8 unique records.

#### 6.1 Label Encoding


```python
# Get the attributes that have equal or more than 8 unique records
label_encoding_col = category_data.columns[category_data.nunique() >= 8]
```

```python
# Check the attributes
predictors[label_encoding_col].head(2)
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

# Convert attribute to strings
predictors[label_encoding_col] = predictors[label_encoding_col].astype(str)

# Store the category features in a variable
category = label_encoding_col

for column in category:
    predictors[column] = label_encoder.fit_transform(predictors[column])
```

```python
#Review the dataset
predictors.head(2)
```

#### 6.2 One Hot Encoding

```python
# Get the attributes that have less than 8 unique records
one_hot_col = category_data.columns[category_data.nunique() < 8]
```

```python
# review the attributes
predictors[one_hot_col].head()
```

```python
predictors_dummies = pd.get_dummies(predictors[one_hot_col], prefix_sep='_', drop_first=True)
```

```python
predictors_dummies.head(2)
```

```python
# Drop the original attributes in the wildfire dataframe
predictors = predictors.drop(columns=one_hot_col)
```

```python
# Combine the dummy variable with the original dataframe wildfire
predictors_combined = pd.concat([predictors, predictors_dummies], axis=1)
```

```python
# Review the dataset
predictors_combined
```

#### 6.3 Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize the dataset
scaler = MinMaxScaler()
scaled = scaler.fit_transform(predictors)
min_max = pd.DataFrame(scaled, columns=predictors.columns)
min_max.head()
```

Combine the predictors and outcome to make a final dataset

```python
# Get the outcome variable
size_class = wildfire['size_class']

# Combine the dataset to make a final dataset
wildfire_final = pd.concat([size_class, predictors_combined], axis=1)
```
